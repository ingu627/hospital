```python
import pandas as pd
import re
import numpy as np
```

```python
hos=pd.read_csv("KOR_hospital.csv")
hos
```

```python
new=hos[["요양기관명","종별코드명","시도코드명","주소","x좌표","y좌표"]] #종별코드 칼럼 빼고 추출
```

```python
cond=(new['종별코드명']=="상급종합")|(new['종별코드명']=="종합병원")|(new['종별코드명']=="병원")
cond

#상급종합병원, 종합병원, 병원 -> True
```

```python
new_df=new[cond] #상급종합병원, 종합병원, 병원 추출
new_df
```

```python
exist_Null=new_df[new_df['x좌표'].isnull()] #x좌표에 결측값이 존재하는 행만 추출
exist_Null
```

```python
exist_Null
```

```python
ro=exist_Null[exist_Null['주소'].str.contains(".+로 \d+[-]*\d*[,]* ")]  # 주소가 '~로' 인 데이터 추출 -> 데이터프레임 ro
gil=exist_Null[exist_Null['주소'].str.contains(".+길 \d+[-]*\d*[,]* ")] # 주소가 '~길' 인 데이터 추출 -> 데이터프레임 gil
```

```python
ro_dict={"index":[], "새주소":[]} #변경된 주소를 담을 딕셔너리

for i in ro.index:
    ro_dict["새주소"].append(re.match(".+로 \d+[-]*\d*",ro["주소"][i])[0])
    ro_dict["index"].append(i)
    
new_ro=pd.DataFrame(ro_dict) #변경된 주소 딕셔너리 -> 데이터프레임 new_ro
new_ro
```

```python
gil_dict={"index":[], "새주소":[]} #변경된 주소를 담을 딕셔너리

for i in gil.index:
    gil_dict["새주소"].append(re.match(".+길 \d+[-]*\d*",gil["주소"][i])[0])
    gil_dict["index"].append(i)
    
new_gil=pd.DataFrame(gil_dict) #변경된 주소 딕셔너리 -> 데이터프레임 new_gil
new_gil
```

```python
#같은 칼럼을 기준으로 merge 하기 위해 인덱스를 초기화해서 기존 인덱스를 index 칼럼으로 가짐
ro=ro.reset_index()
gil=gil.reset_index()
```

```python
fin_ro=ro.merge(new_ro) # 주소가 '~로'인 데이터프레임(변경전, 변경후) merge
fin_gil=gil.merge(new_gil) # 주소가 '~길'인 데이터프레임(변경전, 변경후) merge
```

```python
gilro=pd.concat([fin_ro,fin_gil]) # 로, 길 데이터프레임 합치기
gilro
```

```python
addr_add=gilro[["index","요양기관명","종별코드명","시도코드명","새주소"]] #필요한 칼럼만 추출
addr_add
```

```python
addr_add.rename(columns={'새주소': '주소'}, inplace=True) #새주소 칼럼명을 주소로 변경
addr_add #변경된 주소의 데이터프레임 완성
```

```python
addr_add.to_excel("addr_add.xlsx") #파일로 저장
```

```python
#------------변경된 주소 데이터프레임을 기존 병원 데이터프레임과 합치기-------------
```

```python
addr_add=pd.read_excel("addr_add (1).xlsx")  
                # addr_add (1).xlsx : 구글 스프레드시트의 'geocode by awesome table' 기능을 사용해서 위경도 값 추가한 파일
```

```python
addr_add=addr_add[["요양기관명","종별코드명","시도코드명","주소","y좌표","x좌표"]]
addr_add
```

```python
final=pd.merge(new_df,addr_add,on=["요양기관명","종별코드명","시도코드명"], how="left")
#기존 데이터프레임(new_df)과 변경된주소 데이터프레임(addr_add) merge
final
```

```python
#x좌표_x 가 null이 아니면 원래 있던 값 그대로 가져오고
#           null이면 x좌표_y에 있는 값을 가져와서 'x' 칼럼에 넣음
np.where(pd.notnull(final['x좌표_x']) == True, final['x좌표_x'], final['x좌표_y'])
```

```python
#x좌표_x 가 null이 아니면 원래 있던 값 그대로 가져오고
#           null이면 x좌표_y에 있는 값을 가져와서 'x' 칼럼에 넣음

final["x"]=np.where(pd.notnull(final['x좌표_x']) == True, final['x좌표_x'], final['x좌표_y'])
```

```python
#y좌표_x 가 null이 아니면 원래 있던 값 그대로 가져오고
#           null이면 y좌표_y에 있는 값을 가져와서 'y' 칼럼에 넣음
final["y"]=np.where(pd.notnull(final['y좌표_x']) == True, final['y좌표_x'], final['y좌표_y'])
```
