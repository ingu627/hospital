```python
import pandas as pd
```

```python
fire_engine=pd.read_excel("전국 119안전센터 현황(2018.11.28).xlsx")
fire_engine
```

```python
fire_engine=fire_engine[["시도본부", "소방서","119안전센터명","주소"]] #전화번호 제외한 열 추출
```

```python
sido=fire_engine["시도본부"] #시도본부열 추출
```

```python
for i in range(len(sido)):
    sido[i]=sido[i][:2] #시도본부에서 앞에 두글자만 가져옴
fire_engine["시도본부"]=sido
```

```python
fire_engine
```

```python
import re
```

```python
addr=fire_engine["주소"] #주소 열 추출
```

```python
for i in range(len(addr)):
    a=re.search("[(]",fire_engine['주소'][i]) #괄호가 있는지 확인
    if a != None: #반환값이 있으면(괄호가 있으면)
        addr[i]=addr[i][:a.start()+1] #괄호 앞까지 가져옴

fire_engine["주소"]=addr #주소열에 넣기
```
