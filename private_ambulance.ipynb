```python
import pandas as pd
```

```python
df = pd.read_csv('민간구급차.csv', encoding='cp949', names=['지점명', '구급차종류', '구급차번호', '전화번호', '지역명(1)', '시도구']) ;df
df
```

```python
df=df[["지점명", "구급차종류", "지역명(1)", "시도구"]] #구급차번호, 전화번호 빼고 추출
df.rename(columns ={"지역명(1)":"지역명"}, inplace=True) #지역명(1)칼럼명 변경
df
```

```python
s=df.groupby(["지점명","구급차종류","시도구"],as_index=False).count() #지점별 특수구급차, 일반구급차 수 count
s=s[["지점명", "구급차종류","지역명","시도구"]]
s.rename(columns={"지역명":"구급차수"}, inplace=True)
```

```python
s
```

```python
p=df.groupby(["지점명","지역명","시도구"],as_index=False).count() #지점별 지역명, 시도구 추출
p=p[["지점명","지역명","시도구"]]
p
```

```python
result=pd.merge(s, p, how="left")
result
```

```python
result.to_excel("민간구급차현황_완료.xlsx")
```
